```python
import pandas as pd
import numpy as np 
from textblob import TextBlob
import re
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df.rename(columns={'reviews.text':'Review'},inplace=True)
df.rename(columns={'reviews.title':'Title'},inplace=True)
df.rename(columns={'reviews.rating':'Rating'},inplace=True)
```

```python
df.dropna(subset=['Review','Title'])
```

```python
#working features are reviews.rating  Review Title and id
```

```python
data=pd.DataFrame(df[['Review','Title']])
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data
```

```python
data.drop_duplicates()
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
    
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
    

def getAnalysis(score):
    if score < -0.15:
        return 'Negative'
    elif score > 0.15:
        return 'Positive'
    else:
        return 'Neutral'
     
```

```python
data['Subjectivity']=data['Review'].apply(getSubjectivity)
data['Polarity']=data['Review'].apply(getPolarity)
data['SENTIMENT']=data['Polarity'].apply(getAnalysis)
```
